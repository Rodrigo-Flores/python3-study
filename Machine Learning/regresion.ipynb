```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas.head()
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas)
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, hue="sex", markers=["o", "v"])
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, hue="sex", markers=["o", "v"], scatter_kws={"s":20})
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas)
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, col="sex", hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, col="day")
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, col="day", hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, col="day", row="sex", hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas)
```

```python
sns.lmplot(x="total_bill", y="tip", data=propinas, aspect=1, size=10)
```
