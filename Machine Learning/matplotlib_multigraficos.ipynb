```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
y = x**2
```

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)

graficos[0][0].plot(x,y,"b")
graficos[0][1].plot(x,y,"r")
graficos[1][0].plot(x,y,"y")
graficos[1][1].plot(x,y,"g")

graficos[0][0].set_title("Azul")
graficos[0][1].set_title("Rojo")
graficos[1][0].set_title("Amarillo")
graficos[1][1].set_title("Verde")
```

Tamaño de gráfico

```python
figura = plt.figure(figsize=(10,2))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,"r")
```

Dos gráficos en la mis figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2, label="Al cuadrado")
graficos.plot(x, x**3, label="Al cubo")
graficos.legend(loc=0) # que busque la mejor posicion
```

Color del gráfico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color="r") # Para los coleres se puede colocar el código sexagesimal
```

Linea de gráfico (grosor, transparencia y tipo)

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color="r", linewidth=10)
```

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color="r", linewidth=10, alpha=0.2)
```

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color="r", linewidth=10, alpha=0.2, linestyle=":")
```
