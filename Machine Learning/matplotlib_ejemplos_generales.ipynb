```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y, "red") # "red" = "r"
plt.title("Este es mi gr√°fico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
plt.subplot(1,2,1) # (filas, columnas)
plt.plot(x,y,"g")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,"g")

plt.subplot(2,1,2)
plt.plot(y,x,"b")
```

Programaci√≥n orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.9,0.9]) # (separacion arrriba, separacion abajo, ancho, alto,)
grafico.plot(x,y,"g")
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,"r")

grafico2 = figura.add_axes([0.2,0.5,0.3,0.3])
grafico2.plot(x,y,"b")
```
